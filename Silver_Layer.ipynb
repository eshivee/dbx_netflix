```python
dbutils.widgets.text('filename', 'directors')
param = dbutils.widgets.get('filename')
print(param)
```

```python
df = spark.read.format('csv')\
    .option('header', True)\
    .option('inferSchema', True)\
    .load(f'abfss://bronze@dbxnetflix.dfs.core.windows.net/netflix_{param}/')
```

```python
df.write.format('delta')\
    .mode('append')\
    .option('path', f'abfss://silver@dbxnetflix.dfs.core.windows.net/netflix_{param}')\
    .save()
```
