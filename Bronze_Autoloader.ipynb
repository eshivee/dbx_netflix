```python
spark.read.format('csv').option('inferSchema', True).option('header', True).load('abfss://source@dbxnetflix.dfs.core.windows.net/').display()
```

```python
df = spark.readStream.format("cloudFiles")\
    .option("cloudFiles.format", "csv")\
    .option("cloudFiles.schemaLocation", "abfss://bronze@dbxnetflix.dfs.core.windows.net/checkpoint")\
    .load('abfss://source@dbxnetflix.dfs.core.windows.net/')
```

```python
df.writeStream.format('delta')\
    .option('checkpointLocation', 'abfss://bronze@dbxnetflix.dfs.core.windows.net/checkpoint')\
    .option('path', 'abfss://bronze@dbxnetflix.dfs.core.windows.net/netflix_titles')\
    .trigger(once = True)\
    .start()
```
