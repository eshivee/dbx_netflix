```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
@dlt.table(name = 'gold_netflix_directors')
@dlt.expect_or_drop('rule1' :'show_id is NOT NULL')
def gold_netflix_directors():
  df = spark.readStream.format('delta').load('abfss://silver@dbxnetflix.dfs.core.windows.net/netflix_directors')
  return df
```

```python
@dlt.table(name = 'gold_netflix_category')
@dlt.expect_or_drop('rule1' :'show_id is NOT NULL')
def gold_netflix_category():
  df = spark.readStream.format('delta').load('abfss://silver@dbxnetflix.dfs.core.windows.net/netflix_category')
  return df
```

```python
@dlt.table(name = 'gold_stg_netflix_titles')

def gold_stg_netflix_titles():
    df = spark.readStream.format('delta').load('abfss://silver@dbxnetflix.dfs.core.windows.net/netflix_titles')
    return df
```

```python
@dlt.view(name = 'gold_trns_netflix_titles')

def gold_trns_netflix_titles():
    df = spark.readStream.table('LIVE.gold_stg_netflix_titles')
    df = df.withColumn('new_flag', lit(0))
    return df
```

```python
maindata_rules = {
    'rule1' : 'show_id is NOT NULL',
    'rule2' : 'newflag is NOT NULL'
}
```

```python
@dlt.table(name = 'gold_netflix_tables')
@dlt.exxpect_all_or_drop(maindata_rules)

def gold_netflix_tables():
    df = spark.readStream('LIVE.gold.trns_netflix_titles')
    return df
```
